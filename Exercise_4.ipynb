```python
# Setting features and labels
print('Setting features and labels...')
y = np.array(df['Global_active_power'])

X = np.array(df.drop(['Global_active_power'],1))
```

```python
df
```

```python
# Reshape the input data
X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
# Split data into test and train
print('Split data into test and train...')
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
# Create LSTM model function
def create_model(num_of_neurons,dropout_ratio=0.2,num_of_layers=1):
    # Create LSTM model
    model = Sequential()
    for layer in np.arange(0,num_of_layers):
      model.add(LSTM(units=num_of_neurons,dropout=dropout_ratio,return_sequences=True))
    model.add(Dense(units=1,activation='sigmoid'))
    return model
```

```python
# Create a model
model = create_model(num_of_neurons=100,dropout_ratio=0.1,number_of_layers=4)  
```

```python
# Configuring model
print('Configuring model...')
model.compile(optimizer='adam',loss=mse,metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping
# Configure callback
callback = EarlyStopping(monitor='loss',patience=20)
```

```python
# Training model
print('Training model...')
model.fit(X_train,y_train,epochs=100,batch_size=1000,verbose=1,callbacks=[callback])
```
